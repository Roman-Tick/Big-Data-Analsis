```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("job-market.csv")
df.head()
```

```python
%matplotlib inline
```

```python
%matplotlib inline

#not sure how to make the gradient rainbow colour on the bars
df['Classification'].value_counts(ascending=True).plot(kind='barh', width=0.6, figsize=(14,20), rot=0)
plt.title('Jobs and Number of Postings')
plt.xlabel('Number of Postings')
plt.ylabel('Classification')
```

```python
import matplotlib
#create dataframe for salary ranges

salaryDf = df[df['LowestSalary'].isna()==False]
salaryDf = salaryDf.groupby('LowestSalary').count()
salaryDf['LowestSalary'] = salaryDf.index
salaryDf['Count'] = salaryDf["Id"]
salaryDf.head()
```

```python
#salary ranges plot
```

```python
plt.figure(figsize=(6,6))

labels = salaryDf['LowestSalary']
sizes = df['LowestSalary'].value_counts()
textprops = {"fontsize":12,'color':'black'}

plt.pie(sizes, labels=labels,
    autopct='%.2f%%',
    pctdistance =0.65,
    shadow=False,
    textprops =textprops,
    wedgeprops={'linewidth': 0, 'edgecolor': 'white'},)

centre_circle = plt.Circle((0,0),0.50, fc='white',linewidth=1.00)
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.axis('equal')
plt.title('Job Posts by salary Range')
plt.show()
```
